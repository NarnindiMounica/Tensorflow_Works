### The Hello World of Deep Learning with Neural Networks

 In the case of creating neural networks, one simple case is where it learns the relationship between two numbers. So, for example, if you were writing code for a function like this, you already know the 'rules':

def hw_function(x):
    y = (2 * x) - 1
    return y
    
So how would you train a neural network to do the equivalent task? By using data! By feeding it with a set of x's and y's, it should be able to figure out the relationship between them.

```python
# Importing libraries

import tensorflow as tf
import numpy as np
```

#### Generating the data

Before creating a neural network you will generate the data that you will be working with. In this case, you are taking 6 X's and 6 Y's. You can see that the relationship between these is y=2x-1, so where x=-1, y=-3 etc.

The standard way of declaring model inputs and outputs is to use numpy, a Python library that provides lots of array type data structures. You can specify these values by building numpy arrays with np.array().

```python
# defining model inputs and outputs for training

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
# defining and compiling neural network

model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(units=1)
])

model.summary()
```
